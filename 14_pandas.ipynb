```python
import pandas as pd

# 1. DataFrame: 2D表格
# 2. Series: 1D List

d1 = pd.DataFrame([
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
], columns=["a", "b", "c"])
d2 = pd.DataFrame([
    [11, 12, 13],
    [14, 15, 16],
    [17, 18, 19]
], columns=["a", "b", "e"])
d2
```

```python
pd.concat([d1, d2],
          join="inner",
          axis=0)
```

過濾: 帶入跟資料筆數依樣多的True/False

```python
# 原理
d1[[True, False, True]]
# 你可以利用內建的一些比較來做
fil = d1["c"] > 5
d1[fil]
# and(&) or(|)
fil = (d1["c"] > 5) & (d1["c"] < 8)
d1[fil]
fil = (d1["c"] <= 3) | (d1["c"] >= 9)
d1[fil]
```
